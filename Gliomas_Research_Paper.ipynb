```python
X = data.iloc[:, 1:].values
y = data.iloc[:, 0].values
```

```python
print(y[0:10])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train[:, 1:2] = sc.fit_transform(X_train[:, 1:2])
X_test[:, 1:2] = sc.transform(X_test[:, 1:2])
```

```python
from sklearn.naive_bayes import GaussianNB
bayes = GaussianNB()
bayes.fit(X_train, y_train)
```

```python
y_pred = bayes.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test, y_pred)
```

```python
print(acc)
```

```python
from sklearn.tree import DecisionTreeClassifier
trees = DecisionTreeClassifier()
trees.fit(X_train, y_train)
```

```python
y_pred_trees = trees.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
acc_trees = accuracy_score(y_test, y_pred_trees)
```

```python
print(acc_trees)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_classifier = RandomForestClassifier(n_estimators=20)
rf_classifier.fit(X_train, y_train)
```

```python
y_pred_rf = rf_classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
acc_rf = accuracy_score(y_test, y_pred_rf)
```

```python
print(acc_rf)
```

```python
inputs = keras.Input(shape=(23,), name="Glioma")
layer_1 = layers.Dense(32, activation="relu", kernel_regularizer=l1(0.001), name="dense_layer_1")(inputs)
#drop_out_layer = layers.Dropout(0.5)(layer_1)
#layer_2 = layers.Dense(16, activation="relu", name="dense_layer_2")(layer_1)
outputs = layers.Dense(2, activation="softmax", name="output_layer")(layer_1)
```

```python
model = keras.Model(inputs=inputs, outputs=outputs, name="glioma_model")
```

```python
y_pred_ann = model.predict(X_test)
y_pred_ann = (y_pred_ann > 0.5)
```

```python
model.compile(
    optimizer=keras.optimizers.RMSprop(),
    loss= keras.losses.SparseCategoricalCrossentropy(),
    metrics=[keras.metrics.mae, keras.metrics.sparse_categorical_accuracy]
)
```

```python
model.summary()
```

```python
print("Fit the model on the training data")
history=model.fit(
    X_train,
    y_train,
    batch_size=16,
    epochs=200,
)
```

```python
test_scores = model.evaluate(X_test, y_test, verbose=2)
print("Test loss:", test_scores[0])
print("Test accuracy:", test_scores[1])
dict(zip(model.metrics_names, test_scores))
```
